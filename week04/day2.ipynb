```python
import pandas as pd

sentence = "Pandas makes data analysis easy and powerful"
s = pd.Series([sentence])

# Convert to lowercase
lower = s.str.lower()
print("lower:\n", lower)

# Convert to uppercase
upper = s.str.upper()
print("upper:\n", upper)

# Capitalize first letter
capitalized = s.str.capitalize()
print("capitalized:\n", capitalized)

# Split into words
split = s.str.split()
print("split:\n", split)

# Replace a word
replaced = s.str.replace("data", "information", regex=False)
print("replaced:\n", replaced)

# Find if 'easy' is in the sentence
contains_easy = s.str.contains("easy")
print("contains_easy:\n", contains_easy)

# Get length of the sentence
length = s.str.len()
print("length:\n", length)

# Strip whitespace from both ends
stripped = s.str.strip()
print("stripped:\n", stripped)

# Swap case
swapcase = s.str.swapcase()
print("swapcase:\n", swapcase)

# Title case
title = s.str.title()
print("title:\n", title)

# Find position of 'analysis'
find_analysis = s.str.find("analysis")
print("find_analysis:\n", find_analysis)

# Count occurrences of 'a'
count_a = s.str.count("a")
print("count_a:\n", count_a)

# Pad string with zeros on the left to length 50
zfill = s.str.zfill(50)
print("zfill:\n", zfill)

# Left justify with '*' padding to length 50
ljust = s.str.ljust(50, '*')
print("ljust:\n", ljust)

# Right justify with '*' padding to length 50
rjust = s.str.rjust(50, '*')
print("rjust:\n", rjust)

# Center with '*' padding to length 50
center = s.str.center(50, '*')
print("center:\n", center)

# Remove all vowels
remove_vowels = s.str.replace(r'[aeiouAEIOU]', '', regex=True)
print("remove_vowels:\n", remove_vowels)

# Extract all words
extract_words = s.str.extractall(r'(\w+)')
print("extract_words:\n", extract_words)

# Repeat the string twice
repeated = s.str.repeat(2)
print("repeated:\n", repeated)

# Encode as utf-8
encoded = s.str.encode('utf-8')
print("encoded:\n", encoded)

# Check if string is numeric
isnumeric = s.str.isnumeric()
print("isnumeric:\n", isnumeric)

# Check if string is alpha
isalpha = s.str.isalpha()
print("isalpha:\n", isalpha)

# Check if string is alphanumeric
isalnum = s.str.isalnum()
print("isalnum:\n", isalnum)

# Check if string is lower
islower = s.str.islower()
print("islower:\n", islower)

# Check if string is upper
isupper = s.str.isupper()
print("isupper:\n", isupper)

# Check if string is title case
istitle = s.str.istitle()
print("istitle:\n", istitle)

# Get first 10 characters
first_10 = s.str[:10]
print("first_10:\n", first_10)

# Get last 10 characters
last_10 = s.str[-10:]
print("last_10:\n", last_10)
```
